```python
df.drop(['body'], axis=1, inplace=True)
print(f'size of dataset after removed: {df.shape}')
```

```python
from sklearn.impute import SimpleImputer
```

```python
print(f'no. of null values before imputing: {df.age.isnull().sum()}')
```

```python
imp = SimpleImputer(strategy='mean')
df['age'] = imp.fit_transform(df[['age']])
print(f'no. of null values after imputation: {df.age.isnull().sum()}')
```

```python
def get_parameters(df):
    parameters = {}
    for col in df.columns[df.isnull().any()]:
        if df[col].dtype == 'float64' or df[col].dtype == 'int64' or df[col].dtype == 'int32':
            strategy = 'mean'
        else :
            strategy = 'most_frequent'
            
        missing_values = df[col][df[col].isnull()].values[0]
        parameters[col] = {'missing_values':missing_values, 'strategy':strategy}
    return parameters

get_parameters(df)
```

```python
parameters = get_parameters(df)
```

```python
import numpy as np
```

```python
for col, param in parameters.items():
    missing_values = param['missing_values']
    strategy = param['strategy']

    imp = SimpleImputer(
        missing_values=missing_values,
        strategy=strategy
    )

    df[col] = imp.fit_transform(df[[col]]).ravel()
```

```python
df['family'] = df['sibsp']+df['parch']
df.loc[df['family']>0, 'travel_alone']=0
df.loc[df['family']==0, 'travel_alone']=1
df['travel_alone'].value_counts().plot(title='passenger travelled alone?', kind='bar')
```
